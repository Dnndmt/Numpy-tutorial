Loading an image...

```python
from skimage import io
photo = io.imread('york_minster.jpg')
type(photo)
```

```python
# the size of the image in the browser is defined as 1170 x 714 wich may be lenght x height
photo.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.imshow(photo)
```

Reverse the rows in the image and make the photo upside down by selecting the last - we get an upside down image

```python
plt.imshow(photo[::-1]) # start stop step
```

Now we can select all the rows and reverse the columns to get a mirror image

```python
plt.imshow(photo[:, ::-1])
```

We can only select part of the image...

```python
plt.imshow(photo[100:200, 400:600])
```

Take every other row and every other column to half the size of the image...

```python
plt.imshow(photo[::2, ::2])
```

Applying mathematical functions to a numpy array...

```python
photo
photo_sin = np.sin(photo)
photo_sin
```

```python
func_lst = [np.sum, np.prod, np.mean, np.std, np.var, np.min, np.max, np.argmin, np.argmax]
# since the photo object is an image, we expect a max value of 255 color values
# source: https://stackoverflow.com/questions/20486700/why-we-always-divide-rgb-values-by-255
# argmin gives you the index of the minimum value
# argmax gives you the index of the maximum value

for func in func_lst:
    print('{}:'.format(str(func)), func(photo))
```
